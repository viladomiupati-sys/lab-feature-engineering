**Check for missing values**

```python
spaceship.isnull().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
df_space = spaceship.copy()
df_space = df_space.dropna()
```

```python
df_space.isnull().sum()
```

```python
df_space.head()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
df_space["Cabin"].str[0]  #primero comprobamos que el codigo este bien.
```

```python
df_space["Cabin"] = df_space["Cabin"].str[0] #como el codigo esta bien, ya podemos cambiar la columna. 
```

- Drop PassengerId and Name

```python
df_space = df_space.drop(columns = ["PassengerId", "Name"], axis = 1)
```

```python
df_space.head()
```

- For non-numerical columns, do dummies.

```python
non_numerical = df_space.select_dtypes(include = ['object']).columns
```

```python
df_space = pd.get_dummies(df_space, columns = non_numerical, drop_first = True, dtype = int)
df_space.head()
```

**Perform Train Test Split**

```python
numerical_columns = ["Age", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]

features = df_space[numerical_columns]
target = df_space['Transported']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
y_train.head()
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier(n_neighbors=10) 
knn.fit(X_train_scaled, y_train)
```

- Evaluate your model's performance. Comment it

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = knn.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
